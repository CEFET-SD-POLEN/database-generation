```python
days = np.array([datetime.datetime(2020, 1, 1) + datetime.timedelta(days=i) for i in range(365)])
```

```python
meteorological_columns = ["Tmean mean daily average temperature (ºC).",
"Tmax maximum daily average temperature (ºC).",
"Tmin minimum daily average temperature (ºC).",
"R total daily rainfall (mm).",
"RH daily average relative humidity (%).",
"WS daily average wind speed (km/h).",
"Wind NE daily average frequency of north-easterly winds (%).",
"Wind SE daily average frequency of south-easterly winds (%).",
"Wind SW daily average frequency of south-westerly winds (%).",
"Wind NW daily average frequency of north-westerly winds (%).",
"CF daily average frequency of calms (%).",
"Sunshine daily average sunshine (hours)."]
```

```python
t_max = np.random.random(365)*25 + 10
t_min = t_max - np.random.random(365)*10
```

```python
rainfall = np.random.random(365)*60
```

```python
humidity = np.random.random(365)*100
```

```python
windspeed = np.random.random(365)*100
```

```python
wind = np.random.random((365, 4))
wind = (wind / wind.sum(axis=1)[:,None]) * 100
```

```python
wind_NE = wind[:, 0]
wind_SE = wind[:, 1]
wind_SW = wind[:, 2]
wind_NW = wind[:, 3]
```

```python
cf = np.random.random(365)*100
```

```python
sunshine = np.random.random(365)*6+6
```

```python
t_mean = (t_max + t_min) / 2
```

```python
meteorological = {"datetime":days, "Tmean":t_mean, "Tmax":t_max, "Tmin":t_min, "Rainfall":rainfall, "Humidity":humidity, "WindSpeed":windspeed, "NE":wind_NE, "SE":wind_SE, "SW":wind_SW, "NW":wind_NW, "CalmFrequency":cf, "Sunshine":sunshine}
```

```python
meteorological_dataframe = pd.DataFrame(meteorological).set_index("datetime")
```
